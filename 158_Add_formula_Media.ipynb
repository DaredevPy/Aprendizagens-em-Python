## Adicionando formula em planilha Excel

```python
from openpyxl import Workbook          # Para criar um novo arquivo Excel
from openpyxl import load_workbook     # Para carregar um arquivo Excel existente

Workbook = load_workbook('dados.xlsx') # Carrega o arquivo Excel existente
sheet = Workbook.active                # Seleciona a planilha ativa
  

sheet['D1'] = 'Salario'                # Define o cabeçalho da coluna D
salarios = [1500, 2000, 2500, 3000]    # Lista de salários a serem adicionados
 
for i, salario in enumerate(salarios, start=2): # Inicia a contagem de linhas a partir da linha 2
    sheet[f'D{i}'] = salario                    # Adiciona os salários na coluna D
    
sheet['D6'] = '=AVERAGE(D2:D5)'                 # Calcula a média dos salários na coluna D usando a fórmula AVERAGE do excel
 
Workbook.save('dados_atualizados.xlsx')         # Salva as alterações em um novo arquivo Excel

```
