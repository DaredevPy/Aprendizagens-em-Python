## Leitura a modificação de Células

```python
from openpyxl import load_workbook
from openpyxl import Workbook

workbook = load_workbook('dados.xlsx')
sheet = workbook.active

# Ler o valor da célula
nome = sheet['A2'].value # Pegando o valor da célula A2
idade = sheet['B2'].value # Pegando o valor da célula B2
cidade = sheet['C2'].value # Pegando o valor da célula C2

sheet['B2'] = 30  # Modificando o valor da célula B2

print(f"Nome: {nome}, Idade: {idade}, Cidade: {cidade}")
workbook.save('dados_modificado.xlsx')  # Salvando as alterações em um novo arquivo
```
