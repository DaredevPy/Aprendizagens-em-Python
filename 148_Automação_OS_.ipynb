# Automação em Python com o Módulo OS

## Manipulação de Diretórios

#### Neste modulo vamos exploras Três funções importantes do módulo OS:

_ os.getcwd(): Obtendo o diretório de trabalho atual.


_ os.chdir(path): Mudando o diretório de trabalho.


_ os.listdir(path): Listando arquivos e diretorios.

### os.getcwd(): Obtendo o diretorio de trabalho atual

##### A Função os.getcwd( ) retorna uma string com o caminho absoluto do diretório de trabalho atual.

### Exemplo:

```python
        ### os.getcwd() ###
import os
dir_atual = os.getcwd()
print(f"Diretório atual: {dir_atual}")
# getcwd() retorna o diretório atual de trabalho
```

```python
         ### os.chdir() ###

dir_atual = os.getcwd()
print(f"Diretório atual: {dir_atual}")

novo_dir = os.path.join(dir_atual,r'C:\Users\Python\Desktop\Codigos\Python') # Cria um novo diretório
os.chdir(novo_dir) # Muda o diretório atual para o novo
print(f"Diretorio mudado de {dir_atual} para {novo_dir}")
# chdir() muda o diretório atual de trabalho
print(f"Diretório atual: {dir_atual}")
```

```python
### os.listdir() ###

dir_atual = os.getcwd()
print(dir_atual)

conteudo = os.listdir(dir_atual)

for item in conteudo:
    print(item)
# listdir() lista o conteúdo do diretório atual
```
