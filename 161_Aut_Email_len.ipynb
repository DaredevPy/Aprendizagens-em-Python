## Acessando e Filtrando Emails com imbox

```python
from imbox import Imbox
import json

with open('credenciais_gmail.json', 'r') as file: # Abrindo o arquivo credenciais
    credenciais = json.load(file)                 # Lendo o arquivo com o metodo json.load()
    
    email = credenciais['e-mail']                 # Acessando a informação do e-mail
    senha = credenciais['password']               # Acessando a informação da senha
    servidor = credenciais['host']                # Acessando a informação do servidor 
```

##### Conectando-se ao servidor de email e listando mensagens

```python
with Imbox(servidor, email, senha) as imb:
    mensagens = imb.messages()
    
    # Verificando a quantidade de mensagens recebidas
    print(f'Qtde de mensagens recebidas: {len(mensagens)}')
```
