### Gerando DataFrames dados para os exemplos

DataFrames- Criando estrutura de dataframes para usar os métodos que apenderemos

```python
import pandas as pd
import numpy as np

# Criando um novo dicionário para ser adicionado em um DataFrame
dados = {
    'NomeVendedor': ['Carlos', 'Julia', 'Lucas', 'Fernanda', 'Roberto', 'Camila', 'Ricardo', 'Beatriz', 'Renato', 'Patricia'],
    'QuantidadeVendas': [33, 27, 41, 22, 33, 30, 45, 25, 40, 28],
    'ValorVendido': [156.45, 123.76, 187.45, 34.21, 12, 543, 512, 587, 10, 234],
    'Localidade': ['Sul', 'Nordeste', 'Nordeste', 'Norte', 'Sudeste', 'Sudeste', 'Sudeste', 'Sul', 'Nordeste', 'Centro-Oeste']
}


# Criando novos DataFrames
df = pd.DataFrame(dados)
df
```

Aplicando funções matemáticas em DataFrames

```python
# Criando uma função para muitiplicar o valor vendido por 5%
def comissao(ValorVendido):
    return ValorVendido * 1.05
# Aplicando a função de comissão a coluna 'ValorVendido' 
df['ValorVendido'].apply(comissao)
# .apply() aplica uma função a cada elemento de uma série ou DataFrame

# Adicionando a coluna 'Comissao' ao DataFrame baseada na coluna 'ValorVendido'
# Aplicando a função de comissão na nova coluna 'Comissao'
df['Comissao'] = df['ValorVendido'].apply(comissao)
df
```
