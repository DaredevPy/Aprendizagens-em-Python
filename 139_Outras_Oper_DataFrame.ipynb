# Outras operaçãoes em DataFrame

```python
import pandas as pd
import numpy as np

# Criando um novo dicionário para ser adicionado em um DataFrame
dados = {
    'NomeVendedor': ['Carlos', 'Julia', 'Lucas', 'Fernanda', 'Roberto', 'Camila', 'Ricardo', 'Beatriz', 'Renato', 'Patricia'],
    'QuantidadeVendas': [33, 27, 41, 22, 33, 30, 45, 25, 40, 28],
    'ValorVendido': [156.45, 123.76, 187.45, 34.21, 12, 543, 512, 587, 10, 234],
    'Localidade': ['Sul', 'Nordeste', 'Nordeste', 'Norte', 'Sudeste', 'Sudeste', 'Sudeste', 'Sul', 'Nordeste', 'Centro-Oeste']
}


# Criando novos DataFrames
df = pd.DataFrame(dados)
df
```

## Função de soma .sum()

```python
df['QuantidadeVendas'].sum()
```

# Função de média .mean()

```python
df['QuantidadeVendas'].mean()
```

# Função de produto .product()

```python
# Multiplicando os valores da coluna.
df['QuantidadeVendas'].product()
```

# Função do maior valor .max()

```python
df['QuantidadeVendas'].max()
```

# Função do menor valor .min()

```python
df['QuantidadeVendas'].min()
```

# Função de ordenamento .sort_values()

```python
df.sort_values(by="QuantidadeVendas", ascending=False)
```
