## Backup de Arquivos

###### Neste notebook, vamos criar um script para copiar arquivos de um diretório para um diretório de Backup

```python
import shutil
import os

if not os.path.exists('C:/BackUp'): # Se a pasta de backup não existir...
    os.mkdir('C:/BackUp') # ...cria a pasta de BackUp

origem_dir = 'C:/Origem' # Diretório de origem dos arquivos
backup_dir = 'C:/BackUp' # Diretório de backup

for arquivo in os.listdir(origem_dir): # Percorre todos os arquivos no diretório de origem
    caminho_arquivo = os.path.join(origem_dir, arquivo) # Lista o caminho completo dos arquivos contidos no diretório de origem
    if os.path.isfile(caminho_arquivo): # Verifica se é um arquivo
        shutil.copy(caminho_arquivo, backup_dir) # Copia o arquivo para o diretório de backup
        
print(f"Arquivos copiados de {origem_dir} para {backup_dir}.") # Mensagem de confirmação
```
