```python
import pandas as pd
import numpy as np
```

```python
# Criando um dicionario para ser adicionado em um dataframe
dados = {
    'NomeVendedor': ['Carlos', 'Julia', 'Lucas', 'Fernanda', 'Roberto', 'Camila', 'Ricardo', 'Beatriz', 'Renato', 'Patricia'],
    'QuantidadeVendas': [33, 27, 41, 22, 35, 30, 45, 25, 40, 28],
    'Localidade': ['Sul', 'Nordeste', 'Nordeste', 'Norte', 'Sudeste', 'Sudeste', 'Sudeste','Sul', 'Nordeste','Centro-Oeste']
}
#Criando um DataFrame
df = pd.DataFrame(dados)
print(df)
```

Agrupamento com metodo min()

```python
# Buscando menores vendas e vendedores por localidade
# O proprio pandas ja identifica o campo numerico do dataframe
df.groupby("Localidade").min()
```

```python
# Ordenando o campo QuantidadeVendas de forma crescente
df.groupby("Localidade").min().sort_values(by='QuantidadeVendas', ascending=True)
```
