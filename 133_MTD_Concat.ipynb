```python
import pandas as pd
import numpy as np
```

```python
# Criando um dicionario para ser adicionado em um dataframe
dados1 = {
    'NomeVendedor': ['Carlos', 'Julia', 'Lucas', 'Fernanda', 'Roberto', 'Camila', 'Ricardo', 'Beatriz', 'Renato', 'Patricia'],
    'QuantidadeVendas': [33, 27, 41, 22, 35, 30, 45, 25, 40, 28],
    'Localidade': ['Sul', 'Nordeste', 'Nordeste', 'Norte', 'Sudeste', 'Sudeste', 'Sudeste','Sul', 'Nordeste','Centro-Oeste']
}

dados2 = {
    'NomeVendedor': ['Ana', 'João', 'Marta', 'Felipe', 'Rafael', 'Bianca', 'Rodrigo', 'Isabela', 'Guilherme', 'Talita'],
    'QuantidadeVendas': [23, 37, 31, 42, 35, 40, 25, 45, 30, 38],
    'Localidade': ['Sul', 'Sudeste', 'Nordeste', 'Norte', 'Sudeste', 'Sudeste', 'Nordeste','Sul', 'Centro-Oeste','Centro-Oeste']
}

dados3 = {
    'NomeVendedor': ['Pedro', 'Luiza', 'Gabriel', 'Amanda', 'Bruno', 'Carolina', 'Diogo', 'Eduarda', 'Fabio', 'Gisele'],
    'QuantidadeVendas': [43, 27, 31, 32, 45, 40, 35, 25, 30, 38],
    'Localidade': ['Nordeste', 'Sul', 'Sudeste', 'Norte', 'Sudeste', 'Sudeste', 'Nordeste','Sul', 'Centro-Oeste','Centro-Oeste']
}

dados4 = {
    'NomeVendedor': ['Henrique', 'Igor', 'Julia', 'Kátia', 'Leonardo', 'Mariana', 'Nelson', 'Otavio', 'Patricia', 'Quintino'],
    'QuantidadeVendas': [33, 47, 31, 22, 35, 40, 45, 35, 40, 38],
    'Localidade': ['Sul', 'Sudeste', 'Nordeste', 'Norte', 'Sudeste', 'Sudeste', 'Nordeste','Sul', 'Centro-Oeste','Centro-Oeste']
}
```

```python
# Criando nome para os dataframes
df1 = pd.DataFrame(dados1)
df2 = pd.DataFrame(dados2)
df3 = pd.DataFrame(dados3)
df4 = pd.DataFrame(dados4)
```

Imprimindo os quatro DataFrame em um unico DataFrame

Concatenação basica

```python
# Imprime mas não renumera os indices
pd.concat([df1,df2,df3,df4])
```

```python
# Outra forma de imprimir os data frame
df5 = pd.concat([df1,df2,df3,df4])
print(df5)
```

Concatenação na horizontal usando axis=1

```python
# Imprimindo na Horizontal
df5 = pd.concat([df1,df2,df3,df4], axis=1) # =0 vertical
print(df5)
```

Gerando um unico indice para os quatro dataframe

```python
pd.concat([df1,df2,df3,df4],ignore_index=True)
```
