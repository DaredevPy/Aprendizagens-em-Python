# Manipulação Básica de um Dataframe

```python
import pandas as pd
import numpy as np
df = pd.DataFrame()
dados = {
    'Nome': ['João','Maria','Pedro','Ana','Matheus'],
    'Idade': [23,45,54,12,13,],
    'Cidade': ['Salvador','Belo Horizonte','São Paulo','Porto Alegre','Rio']

}
df = pd.DataFrame(dados)
print(df)
```

```python
# Resultados logicos
df['Idade']>40
```

```python
# Buscando dentro do dataframe idades maiores de 40 anos
# Resultados mais dados 
df[df['Idade']>40] # <> = != # Operadores de Filtragem
```

```python
# Buscando dentro do DataFrame Idade e Cidade específica 
df[ (df['Idade']>30) & (df['Cidade'] == 'Belo Horizonte')]
```
