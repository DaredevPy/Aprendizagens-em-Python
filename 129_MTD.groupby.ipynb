Agrupando dados

```python
import pandas as pd
import numpy as np
```

```python
# Criando um dicionario para ser adicionado em um dataframe
dados = {
    'NomeVendedor': ['Carlos', 'Julia', 'Lucas', 'Fernanda', 'Roberto', 'Camila', 'Ricardo', 'Beatriz', 'Renato', 'Patricia'],
    'QuantidadeVendas': [33, 27, 41, 22, 35, 30, 45, 25, 40, 28],
    'Localidade': ['Sul', 'Nordeste', 'Nordeste', 'Norte', 'Sudeste', 'Sudeste', 'Sudeste','Sul', 'Nordeste','Centro-Oeste']
}
#Criando um DataFrame
df = pd.DataFrame(dados)
print(df)
```

```python
#Agrupando localidades
df.groupby("Localidade")
```

```python
#Somando a quantidade de vendas por localidade
df.groupby('Localidade').sum("QuantidadeVendas")
```

```python
#Somando a quantidade de vendas por localidade de forma decrescente usando ascending com o valor False
df.groupby("Localidade")["QuantidadeVendas"].sum().sort_values(ascending=False)
```
