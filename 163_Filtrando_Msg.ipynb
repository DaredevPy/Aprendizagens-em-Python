## Lendo e filtrando emails

```python
from imbox import Imbox
import json

with open('credenciais_gmail.json', 'r') as file: # Abrindo o arquivo credenciais
    credenciais = json.load(file)                 # Lendo o arquivo com o metodo json.load()
    
    email = credenciais['e-mail']                 # Acessando a informação do e-mail
    senha = credenciais['password']               # Acessando a informação da senha
    servidor = credenciais['host']                # Acessando a informação do servidor 
```

### Exibindo o assunto da mensagem com o metodo subject

```python
with Imbox(servidor, email, senha) as imb:                      # Conectando ao servidor de e-mail
        mensagens = imb.messages()                              # Obtendo as mensagens da caixa de entrada
        
        # Verificando a quantidade de mensagens recebidas
        print(f'Qtde de mensagens recebidas: {len(mensagens)}')
        
        for uid, msg in mensagens:                              # Iterando sobre as mensagens
            print(f"De/: {msg.sent_from[0]['email']}")          # Exibindo o remetente na posição 0 da lista de remetentes
            print(f"Assunto: {msg.subject}")                    # Exibindo o assunto da mensagem com o metodo subject
            print("\n")
            break                                               # Remover este break para listar todas as mensagens
```

### Definindo o remetente para filtrar as mensagens

```python
remetente = 'noreply@glassdoor.com'                               # Definindo o remetente para filtrar as mensagens
with Imbox(servidor, email, senha) as imb:                        # Conectando ao servidor de e-mail
        mensagens = imb.messages(sent_from=remetente)             # Obtendo as mensagens enviadas pelo remetente               
        
        # Verificando a quantidade de mensagens recebidas
        print(f'Qtde de mensagens recebidas: {len(mensagens)}')
        
        for uid, msg in mensagens:                                # Iterando sobre as mensagens
            print(f"De/: {msg.sent_from[0]['email']}")            # Exibindo o remetente na posição 0 da lista de remetentes
            print(f"Assunto: {msg.subject}")                      # Exibindo o assunto da mensagem com o metodo subject
            print(f"Corpo: {msg.body['plain'][0]}")               # Exibindo o corpo da mensagem em texto simples                
            print("\n")
            break                                            
```

### Definindo o destinatário para filtrar as mensagens

```python
destinatario = 'fabiola.sousalucio@hotmail.com'                   # Definindo o destinatário para filtrar as mensagens
with Imbox(servidor, email, senha) as imb:                        # Conectando ao servidor de e-mail  
        mensagens = imb.messages(sent_to=destinatario)            # Obtendo as mensagens enviadas para o destinatário                  
        
        # Verificando a quantidade de mensagens recebidas
        print(f'Qtde de mensagens recebidas: {len(mensagens)}')
        
        for uid, msg in mensagens:                                # Iterando sobre as mensagens
            print(f"De/: {msg.sent_to[0]['email']}")              # Exibindo o destinatário na posição 0 da lista de destinatários
            print(f"Assunto: {msg.subject}")                      # Exibindo o assunto da mensagem com o metodo subject
            print(f"Corpo: {msg.body['plain'][0]}")               # Exibindo o corpo da mensagem em texto simples                
            print("\n")
            break                                            
```
