```python
from faker import Faker
import openpyxl
from openpyxl.styles import Font, Alignment, Border, Side
import random
from datetime import datetime, timedelta

def gerar_dados_compras_xlsx(nome_arquivo="compras_simuladas.xlsx", num_linhas=35):
    """
    Gera um arquivo XLSX com dados simulados de compras.

    Args:
        nome_arquivo (str): O nome do arquivo XLSX a ser criado.
        num_linhas (int): O número de linhas de dados a serem geradas.
    """

    fake = Faker('pt_BR')  # Usar localidade brasileira para nomes e datas
    workbook = openpyxl.Workbook()
    sheet = workbook.active
    sheet.title = "Dados de Compras"

    # Definir cabeçalhos
    headers = ["Categoria do Produto", "Quantidade", "Preço Unitário", "Data da Compra"]
    sheet.append(headers)

    # Estilo para os cabeçalhos
    header_font = Font(bold=True, color="FFFFFF")
    header_fill = openpyxl.styles.PatternFill(start_color="4F81BD", end_color="4F81BD", fill_type="solid")
    header_alignment = Alignment(horizontal="center", vertical="center")
    thin_border = Border(left=Side(style='thin'), right=Side(style='thin'), top=Side(style='thin'), bottom=Side(style='thin'))

    for col_num, header_text in enumerate(headers, 1):
        cell = sheet.cell(row=1, column=col_num)
        cell.font = header_font
        cell.fill = header_fill
        cell.alignment = header_alignment
        cell.border = thin_border
        sheet.column_dimensions[openpyxl.utils.get_column_letter(col_num)].width = 20 # Ajusta largura da coluna

    # Categorias de produtos fixas
    categorias_produtos = ["Eletrônicos", "Alimentos", "Vestuário", "Livros"]

    # Definir o range de datas
    start_date = datetime(2024, 1, 1)
    end_date = datetime(2025, 1, 1)

    for _ in range(num_linhas):
        categoria = random.choice(categorias_produtos)
        quantidade = random.randint(1, 30)
        preco_unitario = round(random.uniform(1.00, 20.00), 2)  # Preço até 20.00 com 2 casas decimais

        # Gerar data aleatória dentro do range
        time_between_dates = end_date - start_date
        random_days = random.randint(0, time_between_dates.days)
        data_compra = start_date + timedelta(days=random_days)

        row_data = [categoria, quantidade, preco_unitario, data_compra.strftime("%Y-%m-%d")] # Formata a data para string

        sheet.append(row_data)

    # Ajustar largura das colunas automaticamente para o conteúdo (opcional, mas útil)
    for column in sheet.columns:
        max_length = 0
        column_letter = openpyxl.utils.get_column_letter(column[0].column)
        for cell in column:
            try:
                if len(str(cell.value)) > max_length:
                    max_length = len(str(cell.value))
            except:
                pass
        adjusted_width = (max_length + 2)
        if adjusted_width > sheet.column_dimensions[column_letter].width: # Apenas se o novo for maior
            sheet.column_dimensions[column_letter].width = adjusted_width

    # Salvar o arquivo
    workbook.save(nome_arquivo)
    print(f"Arquivo '{nome_arquivo}' criado com sucesso com {num_linhas} linhas de dados.")

# Exemplo de uso:
gerar_dados_compras_xlsx()
```
