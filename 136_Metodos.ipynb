```python
import pandas as pd
import numpy as np
```

```python
# Criando um dicionario para ser adicionado em um dataframe
dados = {
    'NomeVendedor': ['Carlos', 'Julia', 'Lucas', 'Fernanda', 'Roberto', 'Camila', 'Ricardo', 'Beatriz', 'Renato', 'Patricia'],
    'QuantidadeVendas': [33, 27, 41, 22, 33, 30, 45, 25, 40, 28],
    'Localidade': ['Sul', 'Nordeste', 'Nordeste', 'Norte', 'Sudeste', 'Sudeste', 'Sudeste','Sul', 'Nordeste','Centro-Oeste']
}
df = pd.DataFrame(dados)
df
```

## Metodo info () ##

```python
# .info() para exibir informações sobre o DataFrame
df.info()
```

## Memory_usage() ##

```python
# memory_usage() para verificar o uso de memória de cada coluna
df.memory_usage()
```

## Metodo unique() ##

```python
# Verifica os valores únicos na coluna 'QuantidadeVendas'
# Eliminando valores duplicados
df['QuantidadeVendas'].unique()  
```

## Metodo nunique() ##


```python
# Conta o número de valores únicos
df['QuantidadeVendas'].nunique()  
```

## Metodo value_counts() ##

```python
# values_counts() para contar a frequência que aparece cada valor
df['QuantidadeVendas'].value_counts()
```
