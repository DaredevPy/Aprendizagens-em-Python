## Anaise de dados 

```python
# Use o Kaggle para gerar o arquivo de dados
# Briefer Dashborads em python

import pandas as pd

# Carregar os dados do arquivo CSV
df = pd.read_csv('campeonato-brasileiro-full.csv')

# Converte a coluna 'data' para o tipo datetime
df['data'] = pd.to_datetime(df['data'], format='%d/%m/%Y', errors='coerce')

# Criar a coluna 'saldo_gols' (diferença entre mandante_placar e visitante_placar)
df['saldo_gols'] = df['mandante_Placar'] - df['visitante_Placar']

# Criar a coluna 'resultado_mandante'(vitoria, Empate ou Derrota )
df['resultado_mandante'] = df.apply(
    lambda row: 'Vitoria' if row['saldo_gols'] > 0 else 'Derrota' if row['saldo_gols'] < 0 else 'Empate',
    axis=1
)

# Exibir as primeiras linhas do DataFrame
print(df.head())
```
