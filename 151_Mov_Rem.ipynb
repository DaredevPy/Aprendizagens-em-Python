## Movimentação e remoção de arquivos e diretórios

##### shutil.move(src, dst): Movendo um arquivo ou diretório.

##### shutill.rmtree(path): Removendo um diretório recursivamente.

```python
#### shutil.move(src, dst): Movendo um arquivo ou diretório.

import shutil
import os
os.chdir('C:\Folder')  # Definindo o diretório de trabalho
with open('arqOrigem1.txt', 'w') as arq: # Criando um arquivo w para escrita
# with open para abrir um arquivo    
    arq.write('python') # write para escrever no arquivo
    
shutil.move('arqOrigem1.txt', 'arqOrigemMovido.txt')   
# Renomeando o arquivo arqOrigem1.txt para arqOrigemMovido.txt

```
