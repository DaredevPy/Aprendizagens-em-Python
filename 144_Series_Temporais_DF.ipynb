# Séries Temporais em DataFrames

```python
# Gerando dados Temporais
import pandas as pd
num_dias =  10
datas = pd.date_range(start='1/1/2025', periods= num_dias)
print(datas)
# Variavel datas recebe um range de datas que inicia em 1/1/2025 e tem 50 dias de duração
# start: data inicial
# periods: chama a variavel datas com o número de dias que queremos(num_dias = 50)
```

# Usando estrutura criada para gerar indice em outro DataFrame

```python
df = pd.DataFrame(range(num_dias), columns=['numeros'], index=datas)
df
# Fazendo um DataFrame com a variavel datas como index e uma coluna numeros que recebe o range de 0 a 9
```

# Tipos de dados do indice

```python
# Acessando o item 8 no dataframe
df.index[8]
```

# Funções para extrair informações

```python
# Acessando o dia do item 8 no dataframe
df.index[9].day
```

```python
# Acessando o mês do item 8 no dataframe
df.index[9].month
```

```python
# Acessando o ano do item 8 no dataframe
df.index[9].year
```

# Filtrando indices de Séries Temporais

```python
# Acessando o dataframe onde o dia é igual a 
df[df.index.day == 8]
```

```python
# Acessando o dataframe onde o ano é 2025
df[df.index.year == 2025]
```
