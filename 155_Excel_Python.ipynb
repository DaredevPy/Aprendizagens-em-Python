## Leitura de arquivo Excel

```python
from openpyxl import load_workbook # Load_workbook para carregar arquivos Excel existentes

# Carregar o arquivo Excel
workbook = load_workbook('dados.xlsx') # Definido o caminho do arquivo Excel
sheet = workbook.active                # Selecionar aba ativa do Excel
dados =[]                              # Lista para armazenar os dados

for linha in sheet.iter_rows(values_only=True): # Iterar sobre as linhas da planilha
    # values_only=True retorna apenas os valores existentes
    dados.append(linha) # Adicionar cada linha lida à lista
dados    
```
