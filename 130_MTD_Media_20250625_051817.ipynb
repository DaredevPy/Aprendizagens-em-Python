Media usando .groupby

```python
import pandas as pd
import numpy as np
```

```python
# Criando um dicionario para ser adicionado em um dataframe
dados = {
    'NomeVendedor': ['Carlos', 'Julia', 'Lucas', 'Fernanda', 'Roberto', 'Camila', 'Ricardo', 'Beatriz', 'Renato', 'Patricia'],
    'QuantidadeVendas': [33, 27, 41, 22, 35, 30, 45, 25, 40, 28],
    'Localidade': ['Sul', 'Nordeste', 'Nordeste', 'Norte', 'Sudeste', 'Sudeste', 'Sudeste','Sul', 'Nordeste','Centro-Oeste']
}
#Criando um DataFrame
df = pd.DataFrame(dados)
print(df)
```

```python
# Media de Vendas por localidade
df.groupby("Localidade")["QuantidadeVendas"].mean()
```

```python
# Pegando a media de vendas e ordenando de forma crescente com 2 casas decimais
df.groupby("Localidade")["QuantidadeVendas"].mean().round(2).sort_values(ascending=True)
```
