### Gerando DataFrames dados para os exemplos

No Pandas, temos uma variedade de métodos que nos permitem importar dados de várias fontes e
exportar nossos dados para outros locais. Esses métodos são essenciais para manipular dados de
forma eficaz em nossos projetos de análise de dados. Iremos aprender os seguintes mátodos:

* read_csv()    - Lê dados de arquivo .csv
* to_csv()      - Grava dados em arquvo .csv
* read_excel()  - Lê dados de arquivo .xlsx
* read_html()   - Lê dados de arquivo .html




Método read_csv()

```python
# Metodo para ler um arquivo CSV com pandas
import pandas as pd
df = pd.read_csv("dados_cartao.csv", sep=',', decimal=".")
df
```

```python
# .info() exibe informações sobre o DataFrame
df.info()
```

Método to_csv()

```python
import pandas as pd

dados = {
    'NomeVendedor': ['Carlos', 'Julia', 'Lucas', 'Fernanda', 'Roberto', 'Camila', 'Ricardo', 'Beatriz', 'Renato', 'Patricia'],
    'QuantidadeVendas': [33, 27, 41, 22, 33, 30, 45, 25, 40, 28],
    'Localidade': ['Sul', 'Nordeste', 'Nordeste', 'Norte', 'Sudeste', 'Sudeste', 'Sudeste', 'Sul', 'Nordeste', 'Centro-Oeste']
}


# Criando novos DataFrames
df = pd.DataFrame(dados)
df
```

```python
# to_csv para salvar o DataFrame em um arquivo CSV
df.to_csv("vendas.csv", sep=",", decimal=".")
```
