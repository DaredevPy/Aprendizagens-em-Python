## Escrita de arquivos Excel

##### _ Escrever e salvar dados em planilha.

```python
from openpyxl import Workbook # Workbook para manipular arquivos Excel

workbook_novo = Workbook() # Criando uma variável tipo Workbook vasia

sheet_nova = workbook_novo.active # Cria uma nova planilha ativa

dados = [
    ['Nome', 'Idade', 'Cidade'],
    ['Alice', 13, 'Ribeirão Preto'], 
    ['Bob', 15, 'São Paulo'],
] 

# Dados a serem adicionados à planilha

for linha in dados: # Itera sobre cada linha de dados
    sheet_nova.append(linha)  # Adicionando dados em cada linha da sheet

workbook_novo.save('Nova_planilha.xlsx')  # Salva a planilha com o nome especificado
```
