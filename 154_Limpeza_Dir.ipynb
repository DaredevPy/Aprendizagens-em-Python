## Limpesa de diretório

##### Neste notebook, vamos criar um script para remover arquivos temporários de um diretório

```python
import os 

temp_dir = 'C:/Users/Python/Desktop/Codigos/Python/temp' # Definindo o diretório temporário
for arquivo in os.listdir(temp_dir):                     # Listando os arquivos da variável temp_dir
    nome_completo = os.path.join(temp_dir, arquivo)      # Acrescentando o caminho ate os arquivos da lista temp_dir
    if os.path.isfile(nome_completo) and arquivo.endswith('.txt'): # Verificando se é um arquivo e se termina com .txt
        os.remove(nome_completo)                         # Remove todos os arquivos .txt do diretório /temp
```
