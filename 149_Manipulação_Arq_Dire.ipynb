# Manipulação de Arquivos e Diretórios

##### os.mkdir(path): Criando um novo diretório.

##### os.makedirs(path): Criando diretório recursivamente.

##### os.remove(path): Removendo um arquivo.

##### os.rmdir(path): Removendo um diretorio vasio.

##### os.rename(src, dst): Renomenando/movendo um arquivo ou diretório
