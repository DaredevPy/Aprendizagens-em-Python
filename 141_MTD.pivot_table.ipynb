# Metodo .pivot_table()

```python
import pandas as pd
import numpy as np

# Criando um novo dicionário para ser adicionado em um DataFrame
dados = {
    'NomeVendedor': ['Carlos', 'Julia', 'Lucas', 'Fernanda', 'Roberto', 'Camila', 'Ricardo', 'Beatriz', 'Renato', 'Patricia'],
    'QuantidadeVendas': [33, 27, 41, 22, 33, 30, 45, 25, 40, 28],
    'ValorVendido': [156.45, 123.76, 187.45, 34.21, 12, 543, 512, 587, 10, 234],
    'Localidade': ['Sul', 'Nordeste', 'Nordeste', 'Norte', 'Sudeste', 'Sudeste', 'Sudeste', 'Sul', 'Nordeste', 'Centro-Oeste']
}


# Criando novos DataFrames
df = pd.DataFrame(dados)
df
```

```python
# NaNs são valores ausentes
pivot_df = df.pivot_table(index='Localidade', columns='NomeVendedor', values='QuantidadeVendas')
print(pivot_df)
```
