## Organização de Arquivos

###### Neste notebook, vamos criar um script para mover arquivos de um diretório para subdiretórios baseados em sua extenção.

```python
import os 
import shutil

diretorio = r'c:\Users\Python\Downloads' # Definindo o diretório a ser organizado
for arquivo in os.listdir(diretorio):    # Listando os arquivos no diretório
    nome_completo_arquivo = os.path.join(diretorio, arquivo) # Combinando o diretório com o nome do arquivo
    if os.path.isfile(nome_completo_arquivo):  # Verificando se é um arquivo
        extensao = arquivo.split('.')[-1]  # Do '.' para a direita, pegando a extensão do arquivo
        extensao_dir = os.path.join(diretorio, extensao)  # Criando o caminho do diretório da extensão
        os.makedirs(extensao_dir, exist_ok=True)  # Criando o diretório se não existir
        shutil.move(nome_completo_arquivo, extensao_dir)  # Movendo o arquivo para o diretório da extensão

print(f'Arquivos organizados em {diretorio}')  # Mensagem de confirmação 
```
