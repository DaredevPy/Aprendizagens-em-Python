## Cópia de Arquivos e Diretórios

##### shutil.copy(src, dst): Copiando um arquivo.

##### shutil.copy2(src, dst): Copiando um arquivo com metadados.

##### shutil.copytree(src, dst): Copiando um diretório recursivamente.

```python
### shutil.copy(src, dst) Copia o arquivo de origem (src) para o destino (dst).

import shutil
import os
#os.mkdir('c:/scripts')  # Criando um diretório chamado 'scripts'
os.chdir('c:/scripts')  # Mudando o diretório atual para 'scripts'

with open('Arquivo.txt', 'w') as arq: # Abrindo um arquivo para escrita e apelidando-o de 'arq'
    arq.write('Eu amo Python!')  # Escrevendo no arquivo 'Arquivo.txt'
    
shutil.copy('Arquivo.txt', 'Arquivo2.txt')  # Copiando o arquivo para um novo arquivo
```

```python
### shutil.copy2(src, dst) Copia o arquivo de origem (src) para o destino (dst), preservando metadados como timestamps.

import shutil
import os
#os.mkdir('c:/scripts')  # Criando um diretório chamado 'scripts'
os.chdir('c:/scripts')  # Mudando o diretório atual para 'scripts'

with open('Arquivo.txt', 'w') as arq: # Abrindo um arquivo para escrita e apelidando-o de 'arq'
    arq.write('Eu amo Python!')  # Escrevendo no arquivo 'Arquivo.txt'
    
shutil.copy2('Arquivo2.txt', 'ArquivoCP2.txt')  # Copiando o arquivo para um novo arquivo
```

```python
### shutil.copytree(src, dst) Copia um diretório inteiro de origem (src) para o destino (dst), incluindo todos os arquivos e subdiretórios.


import shutil
import os

os.makedirs('dir1/dir2/dir3/dir4', exist_ok=True)  # Criando uma estrutura de diretórios aninhados
                                                   # exist_ok=True evita erro se o diretório já existir
with open('arquivorecursivo.txt', 'w') as arq:  
                arq.write('Eu amo Python recursivo!')  # Escrevendo no arquivo 'arquivorecursivo.txt'
shutil.copytree('dir1', 'novo_dir') # Copiando o diretório 'dir1' e todo o seu conteúdo                
                                                       
```
