Método .loc()

```python
import pandas as pd
import numpy as np
df = pd.DataFrame()
dados = {
    'Nome': ['João','Maria','Pedro','Ana','Matheus'],
    'Idade': [23,45,54,12,13,],
    'Cidade': ['Salvador','Belo Horizonte','São Paulo','Porto Alegre','Rio']

}
df = pd.DataFrame(dados)
print(df)
```

Usando o método .loc()

```python
# Buscando o item 0 da lista
df.loc[0]
```

```python
# Buscando mais de um item na lista 
df.loc[[0,1,2]]
```

Usando o método .iloc()

```python
# Buscando o item 0 da coluna 0
df.iloc[0,0]
```

```python
# Buscando item 2 da coluna 2
df.iloc[2,2]
```

```python
# de (1:4) vai pegar 1 até 3 na vertical. o quarto elemento contando do zero
# de (0:2) vai pegar 0 até 1 na horizontal. os dois primeiros numeros contando do zero
# A contagen sempre começa do Zero (0)
df.iloc[1:4,0:2]
```
