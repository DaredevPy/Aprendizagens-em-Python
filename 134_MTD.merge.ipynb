# Merge

DataFrames- Criando estrutura de dataframes para usar os métodos que apenderemos

```python
# DataFrame de funcionários
import pandas as pd

# DataFrame de funcionarios
df_funcionarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniela'],
    'idade': [27, 33, 45, 38],
    'id_departamento': [1, 2, 1, 3]
})

print("DataFrame df_funcionarios:")
print(df_funcionarios)

# DataFrame de departamentos
df_departamentos = pd.DataFrame({
    'nome_departamento': ['RH', 'Vendas', 'Marketing'],
    'id_departamento': [1, 2, 3]
})

print("\nDataFrame df_departamentos:")
print(df_departamentos)

# Repare que a coluna id_departamento esta nos dois dataframes
# Em cima representa os nomes e em baixo o nome dos departamentos
```

## Merge

O método merge() do pandas é usado para combinar dois (ou mais) DataFrames com base em uma coluna (ou um conjunto de colunas) em comum, de maneira semelhante a como as junções (joins) funcionam em SQL.

Exemplo 01:

```python
# Pegando id_departamento dentro do dic. df_departamento e unindo a df_funcionarios
df_merge = df_funcionarios.merge(df_departamentos,on='id_departamento')
print(df_merge)
```

#### Selecionando e imprimindo colunas específicas

```python
# Definindo uma nova variavel: df_final 
# Imprimindo colunas nome, idade, nome_departamento
df_final = df_merge[['nome', 'idade', 'nome_departamento']]
print(df_final)
```

#### Renomeando colunas

```python
# Renomeando coluna nome => nome do funcionario, idade => Idade, nome_departamento => Departamento
df_final =df_final.rename(columns={'nome': 'Nome do Funcionario',
                          'idade': 'Idade',
                          'nome_departamento': 'Departamento'})
print(df_final)

```
